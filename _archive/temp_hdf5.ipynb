```python
source_limited_data_file = '/home/smithdr/ml/elm_data/step_4_shot_limited_data/data_v6.hdf5'
dest_label_file = '/home/smithdr/ml/elm_data/step_5_label_elms/labeled_elms_v3.hdf5'
with (
    h5py.File(source_limited_data_file, 'r') as sfile,
    h5py.File(dest_label_file, 'a') as dfile,
):
    for key, dgroup in dfile['shots'].items():
        assert isinstance(dgroup, h5py.Group)
        sgroup = sfile['shots'][key]
        for tag in ['start_time','stop_time', 'channels']:
            mod_tag = 'bes_'+tag
            dgroup.attrs[mod_tag] = sgroup.attrs[mod_tag]
            if mod_tag in dgroup.attrs and tag in dgroup.attrs:
                del dgroup.attrs[tag]
        for key in sgroup.attrs:
            if key not in dgroup.attrs:
                dgroup.attrs[key] = sgroup.attrs[key]
        if 'with_other_signals' in dgroup.attrs:
            del dgroup.attrs['with_other_signals']
```
